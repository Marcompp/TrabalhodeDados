## Q1 - Manipulando o DataFrame

Crie uma coluna chamada `Hemisfério` baseada na Latitude

A regra de formação é a seguinte:

Valor | Critério
---|---
Norte | Latitude positiva
Sul | Latitude negativa

```python
Hemisfério=[]
for a in range(len(df["Latitude"])):
    if df["Latitude"][a]<0:
        Hemisfério.append('Sul')
    elif df["Latitude"][a]>0:
        Hemisfério.append('Norte')
    else:
        Hemisfério.append('N/A')
df.Hemisfério = Hemisfério
df
```
